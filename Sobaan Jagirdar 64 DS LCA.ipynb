```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['mpg'], y=df['hp'], hue=df['cyl'], palette='viridis', data=df)
plt.title('Scatter Plot with Hue by Cylinders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='mpg', hue='cyl', multiple='stack', bins=10, kde=True)
plt.title('Histogram of MPG by Cylinders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df['drat'], y=df['mpg'], hue=df['cyl'], data=df)
plt.title('Bar Plot of MPG by Rear Axle Ratio and Cylinders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['carb'], y=df['mpg'], hue=df['cyl'], data=df, marker='o')
plt.title('Line Chart of MPG by Carburetor and Cylinders')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['carb'], y=df['mpg'], hue=df['cyl'], data=df, marker='o')
plt.title('Line Chart of MPG by Carburetor and Cylinders')
plt.show()
```
